```python
import numpy as np

```

**Matplot Library**

```python
import matplotlib.pyplot as plt
```

**Set Size of Figure**

```python
plt.figure(figsize=(8, 2))  # Width: 8 inches, Height: 6 inches
```

**Different Type of Plots**

```python
plt.figure(figsize=(8, 2))
                                           # Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)
#----------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                           #bar plot
cateogories=['a','b','c','d']
values=[1,2,3,4]
plt.bar(cateogories, values)
#----------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                          # Histogram
data=np.random.randn(1000)
plt.hist(data,bins=30)
#---------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                          # Scatter Plot
plt.scatter(x,y)
#---------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                          #Box Plot
mean=0
data=[np.random.normal(mean,std,100) for std in range(1,4)]  #100 Samples
plt.boxplot(data)
#---------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                          #Pie chart
labels=['A','B','C','D']
sizes=[15,30,15,60]
plt.pie(sizes,labels=labels,autopct='%1.1F%%')
#---------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                          #Area Plot
x=np.linspace(0,10,100)
y1=x
y2=x**2
plt.fill_between(x,y1,y2,alpha=0.3)
#---------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                          #Heat Map
data1=np.random.rand(10,10)
plt.imshow(data1,cmap='cool',interpolation='nearest')     #hot or cool #None, nearest, bilinear, bicubic,spline16, spline 36, gaussian
#---------------------------------------------------------------------
plt.figure(figsize=(8, 2))
                                          #Violin Plot
mean=0
data=[np.random.normal(mean,std,100) for std in range (1,4)]
x_labels=['a','b','c']
vl=plt.violinplot(data,showmedians=True,vert=False)
plt.yticks([1, 2, 3], x_labels)
plt.legend(vl['bodies'],x_labels)
```

**Title, Legend, Axis Title**

```python
plt.title("My Graph")  # Add a title
plt.xlabel("X-axis")  # Add label for X-axis
plt.ylabel("Y-axis")  # Add label for Y-axis
```

**Show Plot**

```python
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 0
data = [np.random.normal(mean, std, 100) for std in range(1, 4)]

# Custom x-axis tick labels
x_labels = ['Dataset 1', 'Dataset 2', 'Dataset 3']

plt.figure(figsize=(8, 6))
plt.violinplot(data, showmedians=True, vert=False)
plt.title("Violin Plot")
plt.xlabel("Values")
plt.ylabel("Data Sets")

# Set custom x-axis tick labels
plt.yticks([1, 2, 3], x_labels)

plt.show()
```
